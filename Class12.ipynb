# **Practice with Cloud Computing**
Microsoft Copilot Studio is an intuitive, low-code platform that enables users to visually design intelligent solutions like chatbots, automated workflows, and custom Power Apps without complex coding. By leveraging conversational AI and drag-and-drop features, the platform allows organizations to automate customer service, handle repetitive tasks, and streamline data management through seamless integrations with tools like Power Automate and Google Sheets. Ultimately, it serves as a versatile tool for boosting operational efficiency, improving communication, and simplifying process automation across various data sources.

## **Example: Copilot Studio Workflow**
Since Copilot Studio is a low-code, graphical platform managed through Microsoft's cloud interface rather than written in raw code, you don't typically build it using Python. However, we can use Python to replicate exactly what the text describes: a conversational AI chatbot that handles customer service triggers and integrates with data sources. The following example uses a class-based approach to simulate how Copilot Studio processes user intent, triggers actions, and automates a workflow

```python
import datetime

class CopilotStudioSimulator:

    def __init__(self):
        # Simulating an integrated data source:
        self.knowledge_base = {
            'hours': 'We are open Monday to Friday from 9 AM to 6 PM.',
            'refund': 'You can request a refund within 30 days of purchase.',
            'support': 'You can contact human support at support@example.com.',
        }
        # Simulating an automated task log:
        self.workflow_log = []

    def PowerAutomate_Workflow(self, task_name, details):
        '''Simulates an automation flow to handle repetitive tasks.'''
        timestamp = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        log_entry = f'[{timestamp}] WORKFLOW TRIGGERED: {task_name} -> {details}'
        self.workflow_log.append(log_entry)
        print(f'\n⚡ [Power Automate] Executing: {task_name}...')
        print(f'   ↳ Action: {details}')

    def ConversationalAI_Engine(self, user_message):
        '''Interprets natural language and user intent using keywords.'''
        message_lower = user_message.lower()

        # Trigger 1: Frequently Asked Questions:
        if 'hour' in message_lower or 'time' in message_lower:
            return self.knowledge_base['hours']

        elif 'refund' in message_lower or 'return' in message_lower:
            # Automateing a background task while answering:
            self.PowerAutomate_Workflow(
                'Flag Refund Query', 'Logged refund request for review.'
            )
            return self.knowledge_base['refund']

        # Trigger 2: Directing to the responsible department / Human escalation:
        elif 'human' in message_lower or 'agent' in message_lower:
            self.PowerAutomate_Workflow(
                'Escalate Ticket', 'Created support ticket in CRM.'
            )
            return f'I am routing you to a specialist. {self.knowledge_base['support']}'

        # Default fallback if intent isn't recognized:
        else:
            return 'I am sorry, I did not quite catch that. Could you rephrase your question?'


# Running the Simulation:
if __name__ == '__main__':
    # Initializing our low-code copilot simulator:
    copilot = CopilotStudioSimulator()

    print('Copilot Studio Chatbot Active')

    # Scenario A: Simple FAQ lookup:
    print('\nUser: What are your operating hours?')
    response = copilot.ConversationalAI_Engine('What are your operating hours?')
    print(f'Bot: {response}')

    # Scenario B: Triggering an automated workflow:
    print('\nUser: I need a refund for my order.')
    response = copilot.ConversationalAI_Engine('I need a refund for my order.')
    print(f'Bot: {response}')

    # Scenario C: Escalation and routing:
    print('\nUser: Speak to a human agent please.')
    response = copilot.ConversationalAI_Engine('Speak to a human agent please.')
    print(f'Bot: {response}')
```
