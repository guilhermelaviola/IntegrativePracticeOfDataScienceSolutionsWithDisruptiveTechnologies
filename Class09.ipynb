# **Advanced Computer Vision Techniques**
Computer Vision, a branch of Artificial Intelligence, enables computers to interpret images and extract useful information, simulating human vision. It involves capturing images, analyzing them to identify patterns, and utilizing algorithms like Canny Edge Detection to define object contours. Effective training using large, diverse image databases is crucial for accurate pattern recognition. Applications extend beyond edge detection to object recognition, Optical Character Recognition (OCR), and image description. The technology is employed in fields such as archaeology, security, sports, medicine, and more, highlighting its versatility. Platforms like Microsoft Azure and libraries like OpenCV are essential for developing these systems, which complement human intelligence by enhancing decision-making accuracy through ongoing training and adaptation.

```python
# Importing all the necessary libraries and resources:
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Loading the image:
from google.colab.patches import cv2_imshow

img = cv2.imread('hieroglyphs.jpg', cv2.IMREAD_GRAYSCALE)

# Applying the Canny edge detector:
edges = cv2.Canny(img, 100, 200)

# Displaying the image with the edges detected:
cv2_imshow(edges)
```

```python
def detect_circles(image_path):
  try:
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if img is None:
      print(f'Error loading the image: {image_path}')
      return None
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.medianBlur(gray, 5)
    circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 20, param1=50, param2=30, minRadius=10, maxRadius=100)
    if circles is not None:
      circles = np.uint16(np.around(circles))
      for i in circles[0, :]:
        center = (i[0], i[1])
        radius = i[2]
        cv2.circle(img, center, radius, (0, 255, 0), 2) # Green circle
        cv2.circle(img, center, 2, (0, 0, 255), 3) # Red center
      return img
  except Exception as e:
    print(f'Error during the detection: {e}')
    return None

image_with_circles = detect_circles('image_with_circles.jpg')
if image_with_circles is not None:
  cv2_imshow(image_with_circles)
```
