```python
# Importing all the necessary libraries:
import pyvrep
```

```python
# Conecting the simulator:
vrep.connect('127.0.0.1', 19997)

# Getting the robot handle:
robot_handle = vrep.get_object_handle('Sparky')

# Sending a command to move the robot forward:
vrep.set_joint_velocity (robot_handle, 1, 1.0) # Joint 1, speed 1.0

# Disconnecting the simulator:
vrep.disconnect()
```
