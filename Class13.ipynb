## **Practical Activity**
Imagine you work for a large pizza chain. Predicting pizza demand for each day of the week is crucial for optimizing ingredient inventory, avoiding waste, and ensuring efficient customer service. In this activity, we will create an AI model using Keras and TensorFlow to predict pizza demand based on historical data and contextual factors.

```python
url = 'https://gist.githubusercontent.com/denandreychuk/b9aa812f10e4b60368cff69c6384a210/raw/d24f715d9350d674b0b1bf494d82ccdf81de0647/100%2520Sales%2520Records.csv'

# Loading the historical sales data:
df = pd.read_csv(url)
df.head()

# Separating the data into independent (X) and dependent (y) variables:
X = df[['Region', 'Item Type', 'Sales Channel']]
y = df['Order Priority']

# Data normalization:
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Creating and compile the model:
model = Sequential([
    Dense(16, activation='relu', input_shape=(X.shape[1],)),
    Dense(8, activation='relu'),
    Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Dividing the data into training and test sets:
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Training the model:
model.fit(X_train, y_train, epochs=50, batch_size=8, validation_split=0.2)

# Evaluating the model using appropriate metrics:
loss, mse = model.evaluate(X_test, y_test)
print(f'Mean Squared Error: {mse}')

# Using the model to predict the demand for pizzas:
y_pred = model.predict(X_test)
plt.scatter(y_test, y_pred)
plt.xlabel('Real Values')
plt.ylabel('Predicted Values')
plt.show()
```

```python
url = 'https://gist.githubusercontent.com/denandreychuk/b9aa812f10e4b60368cff69c6384a210/raw/d24f715d9350d674b0b1bf494d82ccdf81de0647/100%2520Sales%2520Records.csv'

# Loading the historical sales data:
df = pd.read_csv(url)
print("DataFrame columns:", df.columns) # Debugging line to show actual columns
df.head()

# Separating the data into independent (X) and dependent (y) variables:
# NOTE: Ensure 'Channel' is correctly spelled and exists in df.columns.
# Also, 'Region', 'Item Type', 'Channel', 'Order Priority' are categorical and need encoding.
X = df[['Region', 'Item Type', 'Channel', 'Order Priority']]
y = df['Order Priority'] # Corrected column name for y

# Data preprocessing for categorical features (e.g., One-Hot Encoding)
X = pd.get_dummies(X, columns=['Region', 'Item Type', 'Channel', 'Order Priority'], drop_first=True)

# If y is categorical (like 'Order Priority' often is), it also needs encoding.
# Assuming 'Order Priority' maps to numerical values 'L','M','H','C' -> 0,1,2,3 for simplicity.
# You might need a more sophisticated mapping depending on your data and goal.
# For now, let's just make it numerical by mapping.
# NOTE: This is a simplistic approach; for real-world, consider LabelEncoder or more robust mapping.
priority_mapping = {'L': 0, 'M': 1, 'H': 2, 'C': 3}
# If 'Order Priority' was already numerical, this step would be skipped.
y = y.map(priority_mapping)


# Data normalization:
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Creating and compile the model:
model = Sequential([
    Dense(16, activation='relu', input_shape=(X_scaled.shape[1],)), # Use X_scaled.shape[1] after encoding
    Dense(8, activation='relu'),
    Dense(1, activation='linear') # Output layer for a regression task
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Dividing the data into training and test sets:
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Training the model:
model.fit(X_train, y_train, epochs=50, batch_size=8, validation_split=0.2)

# Evaluating the model using appropriate metrics:
loss, mse = model.evaluate(X_test, y_test)
print(f'Mean Squared Error: {mse}')

# Using the model to predict the demand for pizzas:
y_pred = model.predict(X_test)
plt.scatter(y_test, y_pred)
plt.xlabel('Real Values')
plt.ylabel('Predicted Values')
plt.show()
```
