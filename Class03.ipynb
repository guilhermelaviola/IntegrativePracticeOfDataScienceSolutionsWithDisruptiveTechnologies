# **Introduction to Libraries in Python**
In this lesson, we'll be transported to the 18th century to help Captain Blackbeard organize his chaotic records of loot and treasure using three fundamental Python libraries: NumPy, focused on array structures and efficient mathematical operations; Pandas, which introduces DataFrames for creating and manipulating spreadsheet-like tables; and Matplotlib, used to generate strategic graphs and visualizations. Throughout this journey, we'll learn the concept of libraries as collections of ready-made functions that expand the language, and we'll see practical examples of how to import and apply each of these tools to transform raw data into valuable visual information for the pirate's routes and strategies.

## **Example: Numpy**
NumPy (Numerical Python) is the foundational open-source library for scientific computing and data analysis in Python. It is primarily used to handle large, multi-dimensional arrays and matrices, offering highly optimized, mathematical functions to process these data structures up to 50 times faster than standard Python lists.

```python
import numpy as np

treasures = np.array([1000, 500, 2000, 1500, 800]) # Values ​​in gold coins
average_treasures = np.mean(treasures)
print(f'Average treasures per looting: {average_treasures}')
```

## **Example: Pandas**
Pandas is a popular open-source Python library used for data manipulation and analysis. It provides fast, flexible, and expressive data structures designed to make working with tabular (relational or labeled) data easy and intuitive.

```python
import pandas as pd

travel_data = {'Looting': [1, 2, 3, 4, 5],
               'Coins': treasures,
               'Location':[ 'Lost Island', 'Safe Harbor', 'Treasure Island', 'Costa Brave', 'Paris']}

df = pd.DataFrame(travel_data)
print(df)
```

## **Example: Matplotlib**
Matplotlib is the most popular, open-source data visualization library for the Python programming language. It allows developers and data scientists to turn raw numbers into clear, understandable graphical representations like line charts, bar graphs, scatter plots, and histograms.

```python
import matplotlib.pyplot as plt

plt.bar(df['Looting'], df['Coins'])
plt.xlabel('Looting Number')
plt.ylabel('Gold Coins')
plt.title('Treasures Conquered by Blackbeard')
plt.show()
```
