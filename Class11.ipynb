## **Example: Virtual Environment with Q-learning**
Imagine you are in Neo-Kyoto, a futuristic city where cyber pets are all the rage. You have acquired a mechanical dragon named Sparky, who can learn tricks using machine learning. In this exercise, you will teach Sparky to navigate a virtual environment using the Q-learning algorithm.

```python
# Importing all the necessary libraries and resources:
!pip install numpy gymnasium matplotlib
import numpy as np
import gymnasium as gym
import random
```

```python
env = gym.make('FrozenLake-v1', is_slippery=False)
q_table = np.zeros([env.observation_space.n, env.action_space.n])

learning_rate = 0.1
discount_factor = 0.99
episodes = 1000

for episode in range(episodes):
  state, _ = env.reset()
  done = False

  while not done:
    action = np.argmax(q_table[state]) if random.uniform(0, 1) > 0.1 else env.action_space.sample()
    new_state, reward, done, _, _ = env.step(action)
    q_table[state, action] = (
        (1 - learning_rate) * q_table[state, action] +
        learning_rate * (reward + discount_factor * np.max(q_table[new_state]))
    )
    state = new_state
```

```python
# Evaluate the trained agent
num_eval_episodes = 10
total_rewards = 0

print("Evaluating agent...")
for episode in range(num_eval_episodes):
    state, _ = env.reset()
    done = False
    episode_reward = 0
    print(f"Episode {episode + 1}:")

    while not done:
        action = np.argmax(q_table[state]) # Take greedy action based on learned Q-values
        new_state, reward, done, _, _ = env.step(action)
        episode_reward += reward
        print(f"  State: {state}, Action: {action}, New State: {new_state}, Reward: {reward}, Done: {done}")
        state = new_state

    total_rewards += episode_reward
    print(f"  Episode finished with reward: {episode_reward}\n")

print(f"Average reward over {num_eval_episodes} episodes: {total_rewards / num_eval_episodes}")
env.close()
```
