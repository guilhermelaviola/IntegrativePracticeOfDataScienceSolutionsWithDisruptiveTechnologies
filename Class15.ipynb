## **Example: Ninja To-Do List**
This example shows the backbone of the Ninja To-Do List web application mentioned in your earlier text. It uses the Flask framework to handle two primary actions: showing the list of missions and updating their status.

```python
from flask import Flask, render_template, request, redirect, url_for

app = Flask(__name__)

# List of missions:
missions = []

@app.route('/', methods=['GET', 'POST'])
def index():
  if request.method == 'POST':
    mission = request.form['mission']
    missions.append({'description': mission, 'status': 'In Progress'})
    return redirect(url_for('index'))
  return render_template('index.html', missions=missions)

@app.route('/complete/<int:index>')
def complete (index):
  missions [index] ['status'] = 'Completed'
  return redirect(url_for('index'))

if __name__ == '__main__':
  app.run(debug=True)
```
