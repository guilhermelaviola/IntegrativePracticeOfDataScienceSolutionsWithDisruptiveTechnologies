## **Example: Python Fundamentals**
The following example shows the basic concepts of the Python programming language, covering its syntax, program structure, primitive data types (integers, floating-point numbers, and strings), and arithmetic and logical operators.

```python
# Requesting information from the user:
unit_price = float(input('Enter the unit price of the product: '))
quantity = int(input('Enter the desired quantity: '))
discount = float(input('Enter the discount percentage (0 if there is none): '))

# Calculating the values:
total_price = unit_price * quantity
discount_value = (discount / 100) * total_price
final_price = total_price - discount_value

# Displaying the formatted results:
print('\nPurchase summary:')
print(f'Total price: R$ {total_price:.2f}')
print(f'With discount: R$ {discount_value:.2f} ({discount}%)')
print(f'Final price: R$ {final_price:.2f}')
```
